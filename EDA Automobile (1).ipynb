# Correlation and Causation

#### Correlation:
Correlation is like looking at how two things change together. When one thing goes up, does the other thing also go up, or does it go down? We use correlation to see if there's a connection between two things. If they change together in a predictable way, we say they're correlated. But just because they're correlated doesn't mean one thing causes the other.

#### Causation:
Causation is about figuring out if one thing directly makes another thing happen. It's like understanding if doing one thing actually causes another thing to occur. 

#### Pearson Correlation

The Pearson Correlation **measures the linear dependence between two variables X and Y.**

The resulting coefficient is a value between -1 and 1 inclusive, where:

1: Perfect positive linear correlation.
0: No linear correlation, the two variables most likely do not affect each other.
-1: Perfect negative linear correlation.

- **Pearson Correlation is the default method of the function "corr". Like before, we can calculate the Pearson Correlation of the of the 'int64' or 'float64' variables.**

```python
df.corr()
```

# Significant of the correlation estimate

**When we talk about the significance of a correlation estimate, we're trying to determine whether the observed correlation between two variables is likely to be a real relationship or just a coincidence.**

In statistical terms, the significance of a correlation estimate is determined by conducting **hypothesis testing**. This involves comparing the **observed correlation coefficient (the measure of correlation between two variables)** to a **critical value based** on a chosen level of significance (usually denoted by α, commonly set at 0.05).

**If the observed correlation coefficient is greater than the critical value (in absolute terms), then we conclude that the correlation is statistically significant.** This means that it's unlikely to have occurred due to random chance alone. 

On the other hand, **if the observed correlation coefficient is smaller than the critical value, we conclude that the correlation is not statistically significant.** In other words, the observed relationship between the variables could have occurred by chance.


# P-value

**The P-value tells us if the connection we found between two things is real or just random chance.** 

We usually pick a level of significance, like 0.05, which means we're 95% sure that the connection we found is real. If the P-value is lower than this level, it suggests the connection is real. If it's higher, it suggests the connection might be due to chance. So, the P-value helps us decide how much we can trust the relationship we found between the two things we're studying.
