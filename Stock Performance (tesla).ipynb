url = f'https://query1.finance.yahoo.com/v7/finance/download/{ticker}?period1={period1}&period2={period2}&interval={interval}&events=history&includeAdjustedClose=true'

#### explanation of the body of the URL above
**Base URL**: 
The base URL is https://query1.finance.yahoo.com/v7/finance/download/. This is the main address where the data is hosted.

**Variables within curly braces {}**: 
These are placeholders that will be filled in with actual values when the URL is used. In this case, <code>{ticker}</code>, <code>{period1}</code>, <code>{period2}</code>, and <code>{interval}</code> are placeholders.

**Query Parameters**: 
These are additional pieces of information that are attached to the URL to specify what data you want to retrieve and how you want it formatted. They come after the <code>?</code> in the URL and are separated by <code>&</code>.

**ticker**: 
This is the stock symbol, such as "TSLA" for Tesla.

**period1**:
This represents the start time for the data retrieval, specified in Unix timestamp format (seconds since January 1, 1970).

**period2**: 
This represents the end time for the data retrieval, also specified in Unix timestamp format.
interval: This specifies the frequency of the data, such as <code>1mo</code> for monthly intervals.
events: This parameter specifies what type of events to include in the data (e.g., "history").

**includeAdjustedClose**: This parameter indicates whether to include adjusted close prices in the data.

By formatting the URL string this way, the coder can easily insert the values of <code>{ticker}</code>, <code>{period1}</code>, <code>{period2}</code>, and <code>{interval}</code> into the URL string using Python's f-string formatting. This allows for dynamic creation of the URL based on the variables defined in the code.

**In summary**
the coder edited the URL to make it more flexible and dynamic so that it can be used to retrieve historical financial data for different stocks and time intervals based on the values of the variables <code>{ticker}</code>, <code>{period1}</code>, <code>{period2}</code>, and <code>{interval}</code>.

### Ticker module

The ticker module is like a tool that helps us get information about stocks or companies from the stock market. It as a way to access data and details about a specific company, like its stock price or financial reports. 

 We just need to give it the stock's special letters called the ticker symbol. For example, for Apple, the symbol is **AAPL**.

```python
tesla = yf.Ticker("TSLA")
```

Use Python code with the <code>requests</code> library to download files from URLs.

```python
import requests

# URL of the file to download
url = f'https://query1.finance.yahoo.com/v7/finance/download/{ticker}?period1={period1}&period2={period2}&interval={interval}&events=history&includeAdjustedClose=true'

# Define a user-agent header
headers = {
    'User-Agent': 'Your User Agent'
}

# Send an HTTP GET request to the URL with the user-agent header
response = requests.get(url, headers=headers)

# Check the response status code and handle it accordingly
if response.status_code == 200:
    # Process the response data
    print("Data downloaded successfully!")
else:
    print("Failed to download the data. Status code:", response.status_code)
```

## Part 4 Data Visualization


### 4.1 Visualize the trends for Open, High, Low and Adjusted Close prices over time

- The first subplot displays the trends in Open, High, Low, Close, and Adjusted Close prices over time.

- The second subplot displays the trading volume over time in a line graph.

```python
import matplotlib.pyplot as plt

# Create figure and axis objects
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 10))

# Plotting price data
df[['Open', 'High', 'Low', 'Close', 'Adj Close']].plot(ax=axes[0])
axes[0].set_title('Tesla Stock Prices (2023)')
axes[0].set_xlabel('Date')
axes[0].set_ylabel('Price')

# Plotting volume data
df['Volume'].plot(ax=axes[1], color='purple')
axes[1].set_title('Tesla Stock Volume (2023)')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Volume')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

### 4.2 Identify Monthly Trends

Using the closing price data, analyse and determine if there are any noticeable trends or patterns in the stock prices over the months.

```python
from matplotlib.dates import MonthLocator


# Resample the 'Close' column to monthly frequency and calculate the mean for each month
monthly_close_mean = df['Close'].resample('M').mean()

# Plotting the trends
plt.figure(figsize=(12, 8))

# Plot the monthly closing prices
plt.plot(monthly_close_mean, marker='o', color='red')

plt.title('Monthly Trends in Tesla Stock Closing Prices (2023)')
plt.xlabel('Date')
plt.ylabel('Closing Price in USD')
plt.grid(True)

# Set the x-axis locator to show every month
plt.gca().xaxis.set_major_locator(MonthLocator())

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 4.3 Monthly Volume Analysis

Analyze the volume of trades to identify patterns. High trading volumes during certain periods may indicate significant market interest. Calculate the average trading volume for each month. This can show the liquidity of the stock over time.

```python
# Resample the 'Volume' column to monthly frequency and calculate the mean for each month
monthly_volume_mean = df['Volume'].resample('M').mean()

# Plotting the trends
plt.figure(figsize=(12, 8))

# Plot the monthly average trading volume
plt.bar(monthly_volume_mean.index, monthly_volume_mean, color='blue', alpha=0.7, width=10)

plt.title('Average Monthly Trading Volume of Tesla Stock (2023)')
plt.xlabel('Date')
plt.ylabel('Average Volume')
plt.grid(True)
plt.xticks(rotation=45)

# Set the x-axis locator to show every month
plt.gca().xaxis.set_major_locator(MonthLocator())

plt.tight_layout()
plt.show()
```
